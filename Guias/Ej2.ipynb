```python
estrella = afa[0:35 ,20:60] #Recorte de la estrella

plt.imshow(estrella, cmap = 'Greys')
plt.show()
```

```python
tres_estrellas = np.tile(estrella, 3)
plt.imshow(tres_estrellas, cmap = 'Greys')
plt.show()

print(tres_estrellas.shape)

#relleno con ceros para que el ancho de tres_estrellas y afa coincidan:
altura = tres_estrellas.shape[0]
margen = (afa.shape[1] - tres_estrellas.shape[1])//2

ceros = np.zeros((altura, margen), dtype = int)
tres_estrellas = np.concatenate([ceros, tres_estrellas, ceros], axis = 1)

plt.imshow(tres_estrellas, cmap = 'Greys')
print(tres_estrellas.shape)
```

```python
afa_tres_estrellas = afa.copy()
afa_tres_estrellas[0:35, :] = tres_estrellas

plt.imshow(afa_tres_estrellas, cmap = 'Greys')
print(afa_tres_estrellas.shape)
```
