```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
afa = plt.imread('AFALogo.bmp')
```

```python
plt.imshow(afa, cmap = 'Grays')
plt.show()

print(afa.shape)
print(afa)
```

# a)

```python
afa_cortado = afa[80:156, 20:110]

plt.imshow(afa_cortado, cmap = 'Greys')
plt.show()
```

# b) 

```python
estrellas = afa[0:35, 20:110]
plt.imshow(estrellas, cmap = 'Grays')
plt.show()

estrellas_t = estrellas.T
plt.imshow(estrellas_t, cmap = 'Grays')
plt.show()

print(estrellas)
```

# c)

```python
afa_sin_fondo = np.where(afa == 0, 0, 1)

plt.imshow(afa_sin_fondo, cmap = 'Greys')
plt.show()
```

# d)

```python
afa_fondo_blanco = np.where(afa == 0, 1, afa)

plt.imshow(afa_fondo_blanco, cmap = 'Greys')
plt.show()
```
