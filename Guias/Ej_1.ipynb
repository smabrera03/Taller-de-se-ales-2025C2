```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
pika = plt.imread('pikachu_vs_charmander.jpeg')
```

```python
plt.imshow(pika)
plt.axis('off')

plt.show()

print(pika.shape)
```

```python
zeros = np.zeros_like(pika[:,:,0])
print(pika[:, :, 0].shape)
pika_rojo = np.array([pika[:, :,0], zeros, zeros])

print(pika_rojo.shape)

pika_rojo = np.transpose(pika_rojo, (1, 2, 0))

print(pika_rojo.shape)

plt.imshow(pika_rojo)
plt.axis('off')
plt.show()
```

# a)

```python
pika_min = np.min(pika, axis = 2).astype(int)
pika_max = np.max(pika, axis = 2).astype(int)
pika_grey = 0.5 * (pika_min + pika_max)
print(pika_min.shape)
print(pika_grey.shape)

plt.imshow(pika_grey, cmap = 'Greys')
plt.axis('off')
plt.show()
```

# b)

```python
pika_gris_b = ((1/3) * np.sum(pika, axis = 2))

plt.imshow(255 - pika_gris_b, cmap = 'Greys')
plt.axis('off')
plt.show()
```

# C)

```python
coef = np.array([0.3, 0.59, 0.11])

pika_gris_c = pika @ coef

print(pika_gris_c.shape)

plt.imshow(255 - pika_gris_c, cmap = 'Greys')
plt.axis('off')
plt.show()
```
